```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('sub-data.csv')
df.head()
```

```python
x = 'NETWORTH'
y = 'INCOME'

plt.scatter(x = df[x], y = df[y], s = 5)
plt.title('Total Networth by Total Income')
plt.xlabel('Total Networth by Household')
plt.ylabel('Total Income by Household')
```
