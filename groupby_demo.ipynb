```python
#https://raw.githubusercontent.com/fivethirtyeight/data/master/congress-age/congress-terms.csv
import pandas as pd
import statistics as sts
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
file = 'stuff.csv'
df = pd.read_csv(file)
print(df.head())
```

```python
print(df["age"].value_counts().max)
```

```python
by_device1b_mean = df.groupby('device1b')[('age')].mean()
print(by_device1b_mean)
```

```python
by_party_median = df.groupby('party')[('age')].median()
print(by_party_median)
```

```python
by_party_median.plot(kind = 'bar', title = 'Median Ages by Party', ylabel = 'Median Age', xlabel = 'Party')
```

```python
freq = df.groupby('age')['party'].value_counts()
ax = freq.plot.barh()
```

```python
#Zoom in or else your going to be blind
plt.figure(figsize=[100, 100])
sns.countplot(data=df, x="age")
```
